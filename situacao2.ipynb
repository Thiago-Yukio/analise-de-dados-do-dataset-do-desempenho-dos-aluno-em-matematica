## IMPORTANDO ARQUIVO TRATADO:


```python
#Importando bibliotecas
import pandas as pd
df_tratado = pd.read_csv('data/arquivo_tratado.csv', sep=',')
```

```python
df_tratado.info()
```

## SEPARAÇÃO DOS DADOS PARA TREINAMENTO E TESTE DOS MODELOS:

```python
from sklearn.model_selection import train_test_split

#Separando a base de treino de X e y
X=df_tratado.drop(['reprovacao'],axis=1)
y=df_tratado['reprovacao']

X_treino,X_teste,y_treino,y_teste=train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=42
)
```

```python
lista_parametros_classificacao=[]
lista_parametros_regressao=[]
dict_llm_classificacao={}
dict_llm_regressao={}
```

## MODELOS DE LLMs:

## Modelos de Classificação:

```python
from sklearn import tree

classifica_arv=tree.DecisionTreeClassifier(random_state=42)
classifica_arv=classifica_arv.fit(X_treino,y_treino)
y_pred_arv=classifica_arv.predict(X_teste)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

dict_llm_classificacao['nome']='DecisionTreeClassifier'
accuracy_arv = accuracy_score(y_teste, y_pred_arv)
dict_llm_classificacao['acuracia']=accuracy_arv
print(f'acuracia: {accuracy_arv:.2f}')
precision_arv= precision_score(y_teste, y_pred_arv,average='weighted')
dict_llm_classificacao['precisao']=precision_arv
print(f'precisao: {precision_arv:.2f}')
recall_arv = recall_score(y_teste, y_pred_arv,average='weighted')
dict_llm_classificacao['recall']=recall_arv
print(f'recall: {recall_arv:.2f}')
lista_parametros_classificacao.append(dict_llm_classificacao.copy())
dict_llm_classificacao.clear()
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifica_kne=KNeighborsClassifier(n_neighbors=3)
classifica_kne=classifica_kne.fit(X_treino,y_treino)
y_pred_kne=classifica_kne.predict(X_teste)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

dict_llm_classificacao['nome']='KNeighborsClassifier'
accuracy_kne = accuracy_score(y_teste, y_pred_kne)
dict_llm_classificacao['acuracia']=accuracy_kne
print(f'acuracia: {accuracy_kne:.2f}')
precision_kne= precision_score(y_teste, y_pred_kne,average='weighted')
dict_llm_classificacao['precisao']=precision_kne
print(f'precisao: {precision_kne:.2f}')
recall_kne = recall_score(y_teste, y_pred_kne,average='weighted')
dict_llm_classificacao['recall']=recall_kne
print(f'recall: {recall_kne:.2f}')
lista_parametros_classificacao.append(dict_llm_classificacao.copy())
dict_llm_classificacao.clear()
```

```python
from sklearn.linear_model import LogisticRegression

classifica_log=LogisticRegression(random_state=42)
classifica_log=classifica_log.fit(X_treino,y_treino)
y_pred_log=classifica_log.predict(X_teste)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

dict_llm_classificacao['nome']='KNeighborsClassifier'
accuracy_log = accuracy_score(y_teste, y_pred_log)
dict_llm_classificacao['acuracia']=accuracy_log
print(f'acuracia: {accuracy_log:.2f}')
precision_log= precision_score(y_teste, y_pred_log,average='weighted')
dict_llm_classificacao['precisao']=precision_log
print(f'precisao: {precision_log:.2f}')
recall_log = recall_score(y_teste, y_pred_log,average='weighted')
dict_llm_classificacao['recall']=recall_log
print(f'recall: {recall_log:.2f}')
lista_parametros_classificacao.append(dict_llm_classificacao.copy())
dict_llm_classificacao.clear()
```

## Modelos de Regressão:

```python
from sklearn.linear_model import LinearRegression

classifica_lir=LinearRegression()
classifica_lir=LinearRegression().fit(X_treino,y_treino)
y_pred_lir=classifica_lir.predict(X_teste)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

dict_llm_regressao['nome']='LinearRegression'
r2_lir = r2_score(y_teste, y_pred_lir)
dict_llm_regressao['R2']=r2_lir
print(f'R2: {r2_lir:.2f}')
mae_lir= mean_absolute_error(y_teste, y_pred_lir)
dict_llm_regressao['MAE']=mae_lir
print(f'MAE: {mae_lir:.2f}')
rmse_lir = np.sqrt(mean_squared_error(y_teste, y_pred_lir))
dict_llm_regressao['RMSE']=rmse_lir
print(f'RMSE: {rmse_lir:.2f}')
lista_parametros_regressao.append(dict_llm_regressao.copy())
```

```python
from sklearn.ensemble import RandomForestRegressor

classifica_rfr=RandomForestRegressor()
classifica_rfr=classifica_rfr.fit(X_treino,y_treino)
y_pred_rfr=classifica_rfr.predict(X_teste)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

dict_llm_regressao['nome']='RandomForestRegressor'
r2_rfr = r2_score(y_teste, y_pred_rfr)
dict_llm_regressao['R2']=r2_rfr
print(f'R2: {r2_rfr:.2f}')
mae_rfr= mean_absolute_error(y_teste, y_pred_rfr)
dict_llm_regressao['MAE']=mae_rfr
print(f'MAE: {mae_rfr:.2f}')
rmse_rfr = np.sqrt(mean_squared_error(y_teste, y_pred_rfr))
dict_llm_regressao['RMSE']=rmse_rfr
print(f'RMSE: {rmse_rfr:.2f}')
lista_parametros_regressao.append(dict_llm_regressao.copy())
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

classifica_hgbr=HistGradientBoostingRegressor()
classifica_hgbr=classifica_hgbr.fit(X_treino,y_treino)
y_pred_hgbr=classifica_hgbr.predict(X_teste)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

dict_llm_regressao['nome']='HistGradientBoostingRegressor'
r2_hgbr = r2_score(y_teste, y_pred_hgbr)
dict_llm_regressao['R2']=r2_hgbr
print(f'R2: {r2_hgbr:.2f}')
mae_hgbr= mean_absolute_error(y_teste, y_pred_hgbr)
dict_llm_regressao['MAE']=mae_hgbr
print(f'MAE: {mae_hgbr:.2f}')
rmse_hgbr = np.sqrt(mean_squared_error(y_teste, y_pred_hgbr))
dict_llm_regressao['RMSE']=rmse_hgbr
print(f'RMSE: {rmse_hgbr:.2f}')
lista_parametros_regressao.append(dict_llm_regressao.copy())
```
